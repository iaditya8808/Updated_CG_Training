```python
from transformers import pipeline

# Load LLM
generator = pipeline("text-generation", model="gpt2")

# Sample academic paragraph
article = """
Artificial Intelligence (AI) is transforming modern industries by enabling machines
to perform tasks that previously required human intelligence. In healthcare,
AI algorithms assist in disease diagnosis and drug discovery. In education,
AI-powered systems provide personalized learning experiences. Despite its benefits,
AI also raises concerns regarding job displacement, data privacy, and ethical use.
Future developments in AI will likely focus on improving transparency, fairness,
and collaboration between humans and intelligent systems.
"""

# Prompt for summarization
prompt = f"Summarize the following research paragraph:\n{article}\nSummary:"

# Generate output
result = generator(prompt, max_length=120, num_return_sequences=1)

print(result[0]['generated_text'])
```

```python
prompt_zero = f"""
Summarize the following article in 5 bullet points:

{article}
"""

result = generator(prompt_zero, max_length=150)
print(result[0]['generated_text'])
```

```python
prompt_few = """
Example Article:
Machine learning helps computers learn patterns from data.

Example Summary:
• Machines learn from data
• Improves predictions
• Used in many industries

Example Article:
Deep learning uses neural networks with multiple layers.

Example Summary:
• Neural networks power deep learning
• Effective for image and speech recognition
• Requires large datasets

Now summarize the following article:

""" + article

result = generator(prompt_few, max_length=180)
print(result[0]['generated_text'])
```

```python
prompt_cot = f"""
Analyze the article step by step.

1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
{article}
"""

result = generator(prompt_cot, max_length=200)
print(result[0]['generated_text'])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = """
Artificial Intelligence enables machines to perform tasks
that typically require human intelligence.
"""

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
print("Total Tokens:", len(tokens))
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def research_assistant(article):

    prompt = f"""
    You are an AI research assistant.

    Provide:
    1. Short Summary
    2. Key Insights (3 points)
    3. One actionable recommendation

    Article:
    {article}
    """

    result = generator(prompt, max_length=200)
    return result[0]['generated_text']


# User input
article_input = input("Enter research article:\n")

output = research_assistant(article_input)

print("\nAI Research Assistant Output:\n")
print(output)
```
